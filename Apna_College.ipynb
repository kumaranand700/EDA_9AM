```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate Fibonacci sequence
def fibonacci(n):
    fib_sequence = [0, 1]
    for i in range(2, n):
        fib_sequence.append(fib_sequence[i-1] + fib_sequence[i-2])
    return fib_sequence

# Generate first 100 Fibonacci numbers
fib_numbers = fibonacci(100)

# Create x-axis values (indices)
indices = np.arange(1, 101)

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(indices, fib_numbers, marker='o', linestyle='-', markersize=3)
plt.title('First 100 Fibonacci Numbers')
plt.xlabel('Position in Sequence')
plt.ylabel('Value')
plt.grid(True)
plt.yscale('log')  # Using log scale as Fibonacci numbers grow exponentially

# Add a second subplot to show the beginning of the sequence more clearly
plt.figure(figsize=(12, 6))
plt.plot(indices[:20], fib_numbers[:20], marker='o', linestyle='-', color='purple')
plt.title('First 20 Fibonacci Numbers')
plt.xlabel('Position in Sequence')
plt.ylabel('Value')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Simple Fibonacci program in Python

n = 10  # number of terms to display
first, second = 0, 1

print("Fibonacci sequence up to", n, "terms:")

for i in range(n):
    print(first, end=" ")
    # calculate next term
    next_num = first + second
    first, second = second, next_num
```

```python
import numpy as np

# Create a 2D array (matrix)
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Original 2D array:")
print(arr)

# Accessing elements
print("\nElement at row 1, column 2:", arr[0, 1])   # 2
print("Element at row 2, column 3:", arr[1, 2])   # 6

# Basic operations
print("\nSum of all elements:", np.sum(arr))
print("Row-wise sum:", np.sum(arr, axis=1))
print("Column-wise sum:", np.sum(arr, axis=0))

# Transpose of the array
print("\nTranspose of array:")
print(arr.T)
```

```python
import numpy as np

# 2D array: rows = students, columns = subjects
grades = np.array([[85, 90, 78],   # Student 1
                   [88, 76, 92],   # Student 2
                   [90, 85, 87]])  # Student 3

print("Grades matrix:")
print(grades)

# Calculate average grade for each student (row-wise)
student_avg = np.mean(grades, axis=1)
print("\nAverage grade per student:", student_avg)

# Calculate average grade for each subject (column-wise)
subject_avg = np.mean(grades, axis=0)
print("Average grade per subject:", subject_avg)

# Find the highest grade overall
max_grade = np.max(grades)
print("\nHighest grade in the class:", max_grade)

# Transpose to flip rows/columns
print("\nTranspose (subjects as rows):")
print(grades.T)
```

```python
import pandas as pd

# Create a DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Score": [85, 90, 95]
}

df = pd.DataFrame(data)

print("DataFrame:")
print(df)

# Basic operations
print("\nAverage Age:", df["Age"].mean())
print("Highest Score:", df["Score"].max())
print("\nFilter Age > 28:")
print(df[df["Age"] > 28])
```
