```python
def sum_to_n(n):
    """
    Calculate the sum of integers from 1 to n.
    
    Args:
        n: A non-negative integer
        
    Returns:
        The sum of integers from 1 to n
    """
    if n == 0:
        return 0
    return n + sum_to_n(n-1)
    
# More efficient non-recursive version
def sum_to_n_formula(n):
    """Calculate sum using the arithmetic sequence formula"""
    return n * (n + 1) // 2
```

# Unwinding the Recursion Explained

"Unwinding the recursion" refers to the process that happens after the base case is reached in a recursive function. It's the phase where the function calls that were put on hold start completing in reverse order (last-in, first-out).

## What Happens During Recursion Unwinding

When a recursive function calls itself, the computer:
1. Pauses the current function execution
2. Stores its state (variables, where to resume)
3. Executes the recursive call
4. Once the recursive call completes, it returns to the paused function and continues

Let me explain with our specific example:


```python
def f2(n):
    if n==0:
        return (0)
    print(n)
    c = n+f2(n-1)  # This line pauses execution until f2(n-1) completes
    print(c)
    return c
```


## Visual Representation of Unwinding for f2(5)

1. **During recursive descent** (going down):
   - f2(5) calls f2(4)
   - f2(4) calls f2(3)
   - f2(3) calls f2(2)
   - f2(2) calls f2(1)
   - f2(1) calls f2(0)
   - f2(0) returns 0 immediately (base case)

2. **During unwinding** (coming back up):
   - f2(1) resumes: c = 1+0 = 1, prints 1, returns 1
   - f2(2) resumes: c = 2+1 = 3, prints 3, returns 3
   - f2(3) resumes: c = 3+3 = 6, prints 6, returns 6
   - f2(4) resumes: c = 4+6 = 10, prints 10, returns 10
   - f2(5) resumes: c = 5+10 = 15, prints 15, returns 15

Think of it like a stack of plates - you add plates from bottom to top (recursive calls), then remove them from top to bottom (unwinding). Each function waits for the one above it to complete before it can finish its own execution.

The unwinding process is what allows the results from deeper recursive calls to be used in the calculations of the earlier calls.

## Breaking Down f2(3)

```python
def f2(3):  # n = 3
    if 3==0:  # False
        return (0)
    print(3)  # Prints 3
    c = 3+f2(2)  # This is where execution pauses until f2(2) completes
    # When execution resumes, f2(2) has returned 3
    c = 3+3 = 6  # Now we calculate c
    print(6)  # Prints 6
    return 6  # Returns 6
```


2. So when we go back to f2(3):
   

```python
   c = 3+f2(2)  # f2(2) returns 3
   c = 3+3 = 6  # This is where 3+3 happens
   
```
