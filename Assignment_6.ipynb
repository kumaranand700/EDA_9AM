```python
# If the marks obtained by a student in five different subjects are input through the keyboard, 
# find out the aggregate marks and percentage marks obtained by the student. 
# Assume that the maximum marks that can be obtained by a student in each subject is 100.  

maths = eval(input("Enter marks in Maths :"))
phys = eval(input("Enter marks in Physics :"))
chem = eval(input("Enter marks in Chemistry :"))
bio = eval(input("Enter marks in Biology :"))
eng = eval(input("Enter marks in English :"))

aggregate = maths + phys + chem + bio + eng
percentage = aggregate/5

print(f"Total marks of the Student is {aggregate}")
print(f"Percentage if the Student is {percentage}")

```

```python
#In a town, the percentage of men is 52. The percentage of total literacy is 48.
#If total percentage of literate men is 35 of the total population, 
#write a program to find the total number of illiterate men and women if the population of the town is 80,000

total_population = 80000

total_pop_men = (total_population*52)/100
total_women = total_population - total_pop_men
total_lit = 48
lit_men = 35
lit_women = total_lit - lit_men

total_literate_men = (total_pop_men*35)/100
total_illiterate_men = total_pop_men - total_literate_men
total_literate_women = (total_women *lit_women)/100
total_illiterate_women  = total_women - total_literate_women
print(f"Total illiterare men are : {total_illiterate_men}")
print(f"Total illiterare women are :{total_illiterate_women}")
```
